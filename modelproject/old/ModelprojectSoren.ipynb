```python
import numpy as np
import matplotlib.pyplot as plt

# Set model parameters
n = 0.01         # Population growth rate
s_k = 0.1        # Savings rate for capital
s_h = 0.1        # Saving rate for human capital
g = 0.02         # Technological progress rate
delta = 0.06     # Capital depreciation rate
alpha = 1/3      # Capital share of output
phi = 1/3        # Capital share of human capital
tau = 0.02        # Tax rate on capital
eta = 0.001       # Fraction of tax revenue allocated to human capital investment

# Set initial values
k = 1           # Capital stock
l = 1           # Labour
a = 1           # Technology
h = 1           # Human Capital
y = 1           # Output
s_kt = 0.1      # Savings rate for physical captal
s_ht = 0.1      # Savings rate for human capital
 
# Create empty arrays to store data
k_array = np.empty(1000)
l_array = np.empty(1000)
a_array = np.empty(1000)
h_array = np.empty(1000)
y_array = np.empty(1000)
s_kt_array = np.empty(1000)
s_ht_array = np.empty(1000)

# Simulate the model for 1,000 periods
for t in range(1000):
    k_array[t] = k
    l_array[t] = l
    a_array[t] = a
    h_array[t] = h
    y_array[t] = y
    s_kt_array[t] = s_kt
    s_ht_array[t] = s_ht
    
    k_next = (s_kt*(1-tau) + (1-delta)*k)/(l*a)
    h_next = (s_ht*(1+eta) + (1-delta)*h)/((l*a))
    l_next = l*(1+n)
    a_next = a*(1+g)
    y_next = (k_next**alpha * h_next**phi*(l*a)**(1-alpha-phi))/((l*a))
    s_kt_next = s_kt*y_next
    s_ht_next = s_ht*y_next
    
    k = k_next
    h = h_next
    y = y_next 

# Plot the results
plt.plot(k_array, label='Capital per effective worker')
plt.plot(y_array, label='Output per effective worker')
plt.plot(h_array, label='Human capital per effective worker')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Output per effective worker')
plt.title('Simulation of the Solow model with human capital and tax for 1.000 periods')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

# Define a function to simulate the Solow model
def solow_model(n, s_k, s_h, g, delta, alpha, phi, tau, eta):
    # Set initial values
    k = 1           # Capital stock
    l = 1           # Labour
    a = 1           # Technology
    h = 1           # Human Capital
    y = 1           # Output
    s_kt = s_k      # Savings rate for physical capital
    s_ht = s_h      # Savings rate for human capital
    
    # Create empty arrays to store data
    k_array = np.empty(1000)
    l_array = np.empty(1000)
    a_array = np.empty(1000)
    h_array = np.empty(1000)
    y_array = np.empty(1000)
    s_kt_array = np.empty(1000)
    s_ht_array = np.empty(1000)
    
    # Simulate the model for 1,000 periods
    for t in range(1000):
        k_array[t] = k
        l_array[t] = l
        a_array[t] = a
        h_array[t] = h
        y_array[t] = y
        s_kt_array[t] = s_kt
        s_ht_array[t] = s_ht

        k_next = (s_kt*(1-tau) + (1-delta)*k)/(l*a)
        h_next = (s_ht*(1+eta) + (1-delta)*h)/((l*a))
        l_next = l*(1+n)
        a_next = a*(1+g)
        y_next = (k**alpha * h**phi*(l*a)**(1-alpha-phi))/((l*a))
        s_kt_next = s_kt*y
        s_ht_next = s_ht*y

        k = k_next
        h = h_next
        y = y_next 
        # End of for loop

    # Plot the results
    plt.plot(k_array, label='Capital per effective worker')
    plt.plot(y_array, label='Output per effective worker')
    plt.plot(h_array, label='Human capital per effective worker')
    plt.legend()
    plt.xlabel('Time')
    plt.ylabel('Output per effective worker')
    plt.title('Simulation of the Solow model with human capital and tax for 1,000 periods')
    plt.show()

# Create interactive sliders for the parameters
interact(solow_model, 
         n=(0, 0.05, 0.01), 
         s_k=(0, 1, 0.05), 
         s_h=(0, 1, 0.05), 
         g=(0, 0.05, 0.01), 
         delta=(0, 1, 0.01), 
         alpha=(0, 1, 0.1), 
         phi=(0, 1, 0.1), 
         tau=(0, 1, 0.1), 
         eta=(0, 1, 0.1))
```
